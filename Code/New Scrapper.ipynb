```python
import requests
import bs4 as bs
import pandas as pd
import numpy as np
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:58.0) Gecko/20100101 Firefox/58.0',
    'From': 'data-x@gmail.com' 
}

sections=["mobile","cybercrime-and-digital-threats","cryptocurrency"]
pagenumbers=[12,54,3]
for i in range(len(sections)):
    chars=pd.DataFrame()
    for j in range(1,pagenumbers[i]+1):
        source=requests.get("https://www.trendmicro.com/vinfo/us/security/news/{}/page/{}".format(sections[i],j)).content
        soup=bs.BeautifulSoup(source, features='html.parser')
        main_art=soup.find_all(class_="enclose")
        link=list()
        title=list()
        intro=list()
        for articles in main_art:
            link.append(articles.find('a').get('href'))
            class_title=articles.find(class_='titlelist')
            title.append(class_title.text)
            intro.append(articles.find(class_='blurbEntry').text)
        data=pd.DataFrame(data=link, columns=['Link'])
        data['Title']=title
        data['Intro']=intro
        chars=chars.append(data, ignore_index=True)
        print("page done: ",j)
    chars.to_csv('{}.csv'.format(sections[i]))
    print("Section done: ",sections[i])
    print("\n\n")
    print(chars.head())

    
```

```python
chars=pd.DataFrame()
pgnum=38
for i in range(1,pgnum+1):
    source2=requests.get('https://www.trendmicro.com/vinfo/us/threat-encyclopedia/spam/page/{}'.format(i)).content
    soup2=bs.BeautifulSoup(source2, features='html.parser')
    spams=soup2.find_all(class_='paddingContainer')
    link=list()
    title=list()
    intro=list()
    for articles in spams:
        class_title=articles.find(class_='ContainerListTitle1')
        link.append(class_title.find('a').get('href'))
        title.append(class_title.text)
        intro.append(articles.find('p').text)
    data=pd.DataFrame(data=link, columns=['Link'])
    data['Title']=title
    data['Intro']=intro
    chars=chars.append(data, ignore_index=True)
    print("page done: ",i)
chars.to_csv('email_spam.csv')
print(chars.head())
    
```
