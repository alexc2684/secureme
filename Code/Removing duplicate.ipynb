```python
import numpy as np
import pandas as pd
find=["whatsapp","facebook","android","ios","gmail", "cryptocurrency", "chrome", "instagram","firefox","youtube","windows","mac","linux"]
for x in range(len(find)):
    df=pd.read_csv("{}.csv".format(find[x]))
    link=df['Link']
    title=df['Title']
    intro=df['Intro']
    link2=link
    title2=title
    intro2=intro
    print("Length", len(link2))
    for i in range(len(link2)):
        print("Testing for row:", i)
        if(i==len(link2)):
            break
        for j in range(i+1,len(link2)):
            if(title2[i] == title2[j]):
                link2[j]=""
                title2[j]=""
                intro2[j]=""
            
    df1=pd.DataFrame(data=link2, columns=["Link"])
    df1['Title']=title2
    df1['Intro']=intro2
    df1.replace('',np.nan, inplace=True)
    df1.dropna(inplace=True)
    df1.to_csv("{}2.csv".format(find[x]))
```
