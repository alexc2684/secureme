```python
import requests
import bs4 as bs
import pandas as pd
import numpy as np
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:58.0) Gecko/20100101 Firefox/58.0',
    'From': 'data-x@gmail.com' 
}

source=requests.get("https://thehackernews.com/search/label/tor?updated-max=2013-11-24T05:04:00-11:00&max-results=20&start=35&by-date=false").content
soup=bs.BeautifulSoup(source, features='html.parser')

blogs=soup.find_all(class_='url entry-title page-link')
blog_links=list()
blog_title=list()
category=list()
for x in blogs:
    blog_links.append(x.get('href'))
    blog_title.append(x.text)
    category.append('spying, tor anatomy')
    

data=pd.DataFrame(data=blog_links, columns=['Link'])
data['Blog Title']=blog_title
data['Category']=category
data.to_csv('Blog_details.csv', mode='a', header=False)
data.head(20)
```
