```python
import pandas as pd
import numpy as np
import bs4 as bs
import requests
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:58.0) Gecko/20100101 Firefox/58.0',
    'From': 'data-x@gmail.com' 
}

start="https://www.security.com"
outer_labels=['virus-threats','cybercrime','mobile-wireless','privacy-compliance','security-infrastructure','management-strategy','vendor-directory','scada-ics']
inner_labels=[['vulnerabilities','email-security','virus-malware','white-papers','endpoint-security'],['cyberwarfare','fraud-identity-theft','phishing','malware','tracking-law-enforcement'],['mobile-security','wireless-security'],['audits','compliance','privacy'],['cloud-security','identity-access','data-protection','network-security','application-security'],['risk-management','security-architecture','disaster-recovery','training-certification','incident-response'],[''],['']]

for i in range(len(outer_labels)):
    label1=outer_labels[i]
    length=len(inner_labels[i])
    chars=pd.DataFrame()
    inum=i
    for j in range(length):
        print("Length of Array is: ",length)
        print("j: ",j)
        p=0
        label2=inner_labels[inum][j]
        print("Section: ",label1," in topic: ",label2)
        source=requests.get("https://www.securityweek.com/{}/{}?page={}".format(label1,label2,p)).content
        soup=bs.BeautifulSoup(source, features='html.parser')
        article=soup.find( class_="panel-pane pane-block pane-views-news-industry-block-1")
        page=article.find(class_="item-list")
        pagenumber=page.find(class_="pager-current").text[5:]
        pagenumber=int(pagenumber)
        print(pagenumber," in topic: ",label2)
        
        while(p<pagenumber):
            source2=requests.get("https://www.securityweek.com/{}/{}?page={}".format(label1,label2,p)).content
            soup2=bs.BeautifulSoup(source2, features='html.parser')
            link=list()
            title=list()
            intro=list()
            articles=soup2.find( class_="panel-pane pane-block pane-views-news-industry-block-1")
    
            body=""
            for i in range(1,11):
                if(i==1):
                    body=articles.find(class_="views-row views-row-1 views-row-odd views-row-first")
    
                elif(i==10):
                    body=articles.find(class_="views-row views-row-10 views-row-even views-row-last")
                else:
                    if(i%2==0):
                        body=articles.find(class_="views-row views-row-{} views-row-even".format(i))
                    else:
                        body=articles.find(class_='views-row views-row-{} views-row-odd'.format(i))
                if(body):
                    link_title=body.find('a')
                    link.append("https://www.securityweek.com"+link_title.get('href'))
                    title.append(link_title.text)
                    intr=body.find('div',{'class': "field-content"}).text
                    intro.append(intr)

    
            data=pd.DataFrame(data=link, columns=['Link'])
            data['Title']=title
            data['Intro']=intro
            data['Topic']=label2
            print("Page done:",p)
            chars=chars.append(data, ignore_index=True)
            p=p+1    
    chars.to_csv('{}.csv'.format(label1))
    print("Section ",label1," completed.")       
```
